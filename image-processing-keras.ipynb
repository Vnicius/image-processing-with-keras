## Processamento de imagens

O processamento digital de imagens é o método de realizar operações em imagens de forma a obter uma imagem melhorada ou estilizada, ou para extrair características e informações importantes. Esse é um tipo de processamento de sinais onde a entrada é uma imagem é a saída é pode ser outra imagem, um texto, uma lista de características, um áudio, um vídeo ou outro tipo de sinal.

Para que uma imagem seja digitalizada é necessário que suas informações sejam amostradas e transformadas em matrizes de números, esses números precisam ser quantizados para serem representados digitalmente. Como esse processo pode gerar uma quantidade gigante de dados, também é comum a aplicação de algoritmos de compressão que se aproveitam da redundância de informações para gerarem imagens que sejam mais fáceis de armazenar e transmitir.

## **Keras: Prática 1 - MNIST**



### **Conhecendo a base de dados**

O dataset MNIST *(Modified National Institute of Standards and Technology)* é um conjunto de dados bem conhecido na comunidade, principalmente para introduzir conceitos de aprendizagem de máquina. Ele é composto de **imagens preto e branco dos números de 0 a 9 escritos à mão**. É um dataset voltado para a tarefa de classificação de imagens.

Para a nossa sorte ele é composto de imagens quadradas de **28x28 pixels**, além de conter **60.000** imagens destinadas para o **treinamento** e **10.000** imagens destinadas para os **testes**.

### **Aquisição da base de dados**

O dataset MNIST é tão conhecido e utilizado que ele pode ser facilmente [acessado com o Keras](https://keras.io/api/datasets/mnist/).

````
from tensorflow.keras.datasets import mnist

# Carregando o dataset
# O x corresponde aos nossos dados de entrada (as imagens)
# O y corresponde aos nossos dados de saída (0-9)
(x_train, y_train), (x_test, y_test) = mnist.load_data()

print("X Train: ", x_train.shape)
print("Y Train: ", y_train.shape)
print("X Test: ", x_test.shape)
print("Y Test: ", y_test.shape)
````

### **Preparando os dados**

Antes dos nosso serem enviada para a rede neural são necessários realizar alguns procedimentos como normalização e *encoding* dos dados. Muitos outros processos podem ser realizar para termos uma base de dados concisa, mas esse dataset em especial não irá precisar deles.


````
import numpy as np

NUM_CLASSES = 10   # os valores variam de 0 a 9
DATA_HEIGHT = 28  # as imagens possuem 28x28 pixels
DATA_WIDTH = 28
````

#### **Normalização**

Vamos normalizar os dados para diminuir a variação de valores. Como bem sabemos, os valores das cores variam de 0 a 255, uma variação de valores tão grande pode atrapalhar no aprendizado da rede, o ideal é que essa variação seja a menor possível, esse é o processo de normalização. E o que melhor para isso do que valores decimais?

Vamos fazer com que os valores variem entre 0 e 1!

````
x_train = x_train.astype("float32") / 255
x_test = x_test.astype("float32") / 255

# Para que essas imagens sejam aceitos pela rede neural
# é necessário que elas possuam o shape (28, 28, 1)

x_train = np.expand_dims(x_train, -1)
x_test = np.expand_dims(x_test, -1)

print("X Train: ", x_train.shape)
print("X Test: ", x_test.shape)
````

#### **Encoding**

Por fim, iremos realizar um processo de encoding das respostas. As respostas da rede neural não serão os valores exatos de 0 a 9, e sim os valores do neurônios para a dada função de ativação.

Como estamos trabalhando com uma tarefa de classificação, é muito comum realizar o procedimento de *one-hot encoding*, onde cada classe do problema será mapeada para um vetor distinto, nesse *encoding* em particular, cada classe é mapeada para um vetor de valores binários (0 ou 1) onde apenas um dos elementos contém o valor 1. Por exemplo:

```
  Classe 1 -> [1, 0, 0, 0]
  Classe 2 -> [0, 1, 0, 0]
  Classe 3 -> [0, 0, 1, 0]
  Classe 4 -> [0, 0, 0, 1]
```

Para a rede neural independe se essas classes são números, letras, palavras, frases ou outros vetores, contanto que você faça o encoding. Posteriormente você deverá mapear os vetores de volta para as classes reais.

O Keras possibilita que nós façamos isso de uma forma bem simples.

````
from tensorflow.keras.utils import to_categorical

y_train = to_categorical(y_train, NUM_CLASSES)
y_test = to_categorical(y_test, NUM_CLASSES)

print("Y Train: ", y_train.shape)
print("Y Test: ", y_test.shape)
````

### **Camadas**

Para esse exemplos nós utilizaremos os seguintes tipos de camadas:

<p align="center">
<img src="https://imgur.com/Tp3vX7l.png" width="70%"/>
</p>

- [**Dense**](https://keras.io/api/layers/core_layers/dense/) - camada densa com os neurônios totalmente conectados;
- [**Dropout**](https://keras.io/api/layers/regularization_layers/dropout/) - camada responsável por realizar o processo de *dropout* na rede;
- [**Conv2D**](https://keras.io/api/layers/convolution_layers/convolution2d/) - camada convolucional para dados em 2 dimensões;
- [**MaxPooling2D**](https://keras.io/api/layers/pooling_layers/max_pooling2d/) - camada responsável por realizar o processo de *max pooling* em dados 2D.
- [**Flatten**](https://keras.io/api/layers/reshaping_layers/flatten/) - camada que realiza o processo de *flatten* e organizar todos os dados em uma única dimensão.

````
from tensorflow.keras.layers import Dense, Dropout, Conv2D, MaxPooling2D, Flatten, Input
````

### **Montando o modelo**

````
from tensorflow.keras.models import Sequential

CHANNELS = 1
INPUT_SHAPE = (DATA_HEIGHT, DATA_WIDTH, CHANNELS)

# Iremos construir uma rede com camadas em sequência
model = Sequential()
model.add(Input(shape=INPUT_SHAPE))

# Vamos adicionar duas camada convolucionais e duas de max pooling
model.add(Conv2D(32, kernel_size=(5, 5), activation="relu"))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, kernel_size=(5, 5), activation="relu"))
model.add(MaxPooling2D(pool_size=(2, 2)))

# Vamos adicionar uma camada de Flatten para que a saída da camada convolucional
# seja utilizada na rede densa
model.add(Flatten())

model.add(Dropout(0.5))

# Essa será a nossa camada de saída
model.add(Dense(NUM_CLASSES, activation="softmax"))

model.summary()
````

### **Treinando o modelo**

````
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

def plot_history(history):
  plt.rcParams["figure.figsize"] = (20,6)
  fig, (acc_plt, loss_plt) = plt.subplots(1, 2)
  
  acc_plt.plot(history.history['accuracy'])
  acc_plt.plot(history.history['val_accuracy'])
  acc_plt.set_title('model accuracy')
  acc_plt.set(xlabel='epoch', ylabel='accuracy')
  acc_plt.legend(['train', 'valid'], loc='upper left')

  loss_plt.plot(history.history['loss'])
  loss_plt.plot(history.history['val_loss'])
  loss_plt.set_title('model loss')
  loss_plt.set(xlabel='epoch', ylabel='loss')
  loss_plt.legend(['train', 'valid'], loc='upper left')

  fig.show()
````

````
from sklearn.metrics import confusion_matrix
import seaborn as sns
import numpy as np

def plot_confusion_matrix(y_true, y_pred):
  cf_matrix = confusion_matrix(y_true = y_true, y_pred = y_pred)
  sns.heatmap(cf_matrix, annot=True, cmap='Blues')
````

````
BATCH_SIZE = 128
EPOCHS = 15

model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

history = model.fit(x_train, y_train, batch_size=BATCH_SIZE, epochs=EPOCHS, validation_split=0.1)
````

````
plot_history(history)
````

### **Avaliando o modelo**


````
score = model.evaluate(x_test, y_test, verbose=0)
print("Loss:", score[0])
print("Accuracy:", score[1])
````

````
predictions = model.predict_classes(x_test)

plot_confusion_matrix(np.argmax(y_test, axis=1), predictions)
````
